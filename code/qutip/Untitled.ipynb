```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
```

$α = cos({θ \over 2})$\
$β = e^{iΦ}sin({θ \over 2})$

$α = {1 \over \sqrt2}$\
$β = {1 \over \sqrt2}$

${1 \over \sqrt2} = cos({θ \over 2})$\
$θ = {\pi \over 2}$

${1 \over \sqrt2} = e^{iΦ}sin({θ \over 2})$\
$θ = {\pi \over 2}$\
$Φ = 0$


```python
def get_theta(n):
    return 2 * np.arcsin(n)

def get_points(theta, phi):
    x = np.cos(phi)*np.sin(theta)
    y = np.sin(phi)*np.sin(theta)
    z = np.cos(theta)

    return [ x, y, z ]
```

```python
b = qutip.Bloch()

n = 1/np.sqrt(2)

theta = get_theta(n)
phi = 0

b.add_vectors(get_points(theta, phi))

b.show()
```

```python
b = qutip.Bloch()
phi = 0

vectors = [get_points(get_theta(theta), phi) for theta in [0, 1/np.sqrt(2), 1/np.sqrt(3), 1, -1/np.sqrt(2)]]
display(vectors)

b.add_vectors(vectors)

b.show()
```

```python
b = qutip.Bloch()
phi = np.pi/2

vectors = [get_points(get_theta(theta), phi) for theta in [0, 1/np.sqrt(2), 1/np.sqrt(3), 1, -1/np.sqrt(2)]]
display(vectors)

b.add_vectors(vectors)

b.show()
```
