```python
from qiskit import QuantumCircuit, BasicAer, execute
from qiskit.quantum_info import Statevector
```

```python
encoded_input = QuantumCircuit(2, 2)

encoded_input.x(0)
encoded_input.x(1)

encoded_input.draw("mpl")
```

```python
transformation = QuantumCircuit(2, 2) 
transformation = encoded_input.compose(transformation) 

transformation.barrier()

#transformation.h(range(2)) # it inverts the control and target qubit
transformation.cx(0, 1)
#transformation.h(range(2))

transformation.draw("mpl")
```

```python
st = Statevector(transformation)
st.draw("latex")
```

```python
circuit1 = QuantumCircuit(3)


circuit1.x(range(3))
circuit1.barrier()
circuit1.cx(0, 2)


circuit1.draw("mpl")
```

```python
st = Statevector(circuit1)
st.draw("latex")
```

```python
circuit2 = QuantumCircuit(3)

circuit2.x(range(3))

circuit2.barrier()

circuit2.cx(1, 2)
circuit2.cx(0, 1)
circuit2.cx(1, 2)
circuit2.cx(0, 1)



circuit2.draw("mpl")
```

```python
st = Statevector(circuit2)
st.draw("latex")
```
