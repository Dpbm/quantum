```python
from qiskit import QuantumCircuit, Aer
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
```

```python
qc_1 = QuantumCircuit(2, 1)

qc_1.x(-1)
qc_1.h(range(2))

st0 = Statevector.from_instruction(qc_1)

qc_1.barrier()
qc_1.cx(0, 1)
qc_1.x(0)
st1 = Statevector.from_instruction(qc_1)

qc_1.barrier()


qc_1.h(0)

st2 = Statevector.from_instruction(qc_1)
display(qc_1.draw("mpl", fold=-1))
display(Statevector(st0).draw("latex"))
display(Statevector(st1).draw("latex"))
display(Statevector(st2).draw("latex"))

```

```python
qc_1.measure(0, 0)

sim = Aer.get_backend("aer_simulator")
results = sim.run(qc_1).result().get_counts()
plot_histogram(results)
```

```python
qc_2 = QuantumCircuit(3, 2)

qc_2.x(-1)
qc_2.h(range(3))

st0 = Statevector.from_instruction(qc_2)

qc_2.barrier()
#qc_2.cx(0, 1)
#qc_2.cx(0, 2)
qc_2.x(0)
st2 = Statevector.from_instruction(qc_2)

qc_2.barrier()


qc_2.h(range(2))

st2 = Statevector.from_instruction(qc_2)
display(qc_2.draw("mpl", fold=-1))
display(Statevector(st0).draw("latex"))
display(Statevector(st1).draw("latex"))
display(Statevector(st2).draw("latex"))

```

```python
qc_2.measure(range(2), range(2))

sim = Aer.get_backend("aer_simulator")
results = sim.run(qc_2).result().get_counts()
plot_histogram(results)
```
