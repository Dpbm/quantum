```python
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
from itertools import product
```

```python
sim = Aer.get_backend("aer_simulator")

for i in product('01', repeat=2):
    bit_string = ''.join(i)

    circuit = QuantumCircuit(2, 1)
    
    for qubit, bit in enumerate(bit_string[::-1]):
        if(bit == '1'):
            circuit.x(qubit)
    
    circuit.barrier()
    
    circuit.x(0)
    circuit.h(range(2))
    
    circuit.barrier()
    circuit.cx(1, 0)
    circuit.barrier()
    
    circuit.h(range(2))
    
    display(Statevector(circuit).draw('latex'))
    
    circuit.measure(1, 0)
    display(circuit.draw('mpl'))
    
    
    result = sim.run(transpile(circuit, sim), shots=1000).result().get_counts()
    display(plot_histogram(result))
    
    
    if(result.get('0')):
        print("they aren't equal")
    else:
        print("They are the same")
    
    
    print('-'*100)
    
    
    
```
