```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, Aer
from qiskit.quantum_info import Statevector
```

```python
unknown_state = QuantumRegister(1, name="Ïˆ")
known_states = QuantumRegister(2)
out = ClassicalRegister(3)
```

```python
qc = QuantumCircuit(known_states,unknown_state, out)

qc.h(unknown_state)
qc.barrier()

qc.cnot(unknown_state, known_states[1])
qc.cnot(unknown_state, known_states[0])

qc.draw("mpl")
```

```python
Statevector(qc).draw("latex")
```

```python
qc.measure(unknown_state, out[0])
qc.measure(known_states, out[1:])
qc.draw("mpl")
```

```python
sim = Aer.get_backend("aer_simulator")
sim.run(qc).result().get_counts()
```
