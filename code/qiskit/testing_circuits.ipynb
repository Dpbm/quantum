```python
from qiskit import QuantumCircuit
import numpy as np
```

```python
qc = QuantumCircuit(3)

qc.h(range(2))

#qc.h(0)
#qc.z(0)
#qc.h(1)
#qc.z(1)
qc.z(2)
qc.rz(np.pi, 2)

qc.draw("mpl")
```

```python
from qiskit.quantum_info import Statevector
st = Statevector(qc)
st.draw("latex")
```

```python
from qiskit.visualization import plot_bloch_multivector
plot_bloch_multivector(st)
```
