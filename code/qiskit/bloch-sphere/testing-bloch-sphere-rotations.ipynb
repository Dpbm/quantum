```python
plot_points = []
for theta in range(360):
    RY = np.matrix([
        [ np.cos(theta/2), -np.sin(theta/2) ],
        [ np.sin(theta/2), np.cos(theta/2)  ]
    ])
    zero = np.matrix([
        [1],
        [0]
    ])
    
    result_matrix = RY * zero
    result_list = result_matrix.tolist()
    plot_points.append([result_list[0][0], result_list[1][0]])
```

```python
plot_points
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(plot_points)
```

```python
x = y = z = []
for theta in range(360):
    for psi in range(180):
        x.append(np.sin(psi)*np.cos(theta))
        y.append(np.sin(psi)*np.sin(theta))
        z.append(np.cos(psi))
```

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection ='3d')
ax.plot3D(x, y, z, 'green')
plt.show()
```
