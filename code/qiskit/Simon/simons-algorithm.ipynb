```python
from qiskit import QuantumCircuit, Aer
from qiskit.quantum_info import Statevector
import numpy as np
```

```python
sim = Aer.get_backend('aer_simulator')
```

```python
test = QuantumCircuit(3, 2)
test.h(range(2))
#test.cx(0, 2)
test.cx(1, 2)
test.h(range(2))
test.measure(range(2), range(2))

display(test.draw('mpl'))
sim.run(test).result().get_counts()
```

```python
test = QuantumCircuit(4, 2)
test.h(range(2))

test.barrier()
test.cx(0, 2)
test.cx(1, 3)
test.cx(1, 3)
test.swap(2,3)
test.barrier()


test.h(range(2))
test.measure(range(2), range(2))

display(test.draw('mpl'))
sim.run(test).result().get_counts()
```
