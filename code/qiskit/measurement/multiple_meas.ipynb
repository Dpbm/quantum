```python
%matplotlib inline
from qiskit import QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator
```

```python
sim = AerSimulator()
plot = lambda qc: plot_histogram(sim.run(qc, shots=1000).result().get_counts())
```

```python
qc = QuantumCircuit(1,1)
qc.measure(0, 0)
qc.x(0)
qc.measure(0, 0)
qc.x(0)
qc.measure(0, 0)
qc.draw('mpl')
```

```python
plot(qc)
```

```python
qc = QuantumCircuit(1,1)
qc.x(0)
qc.h(0)
qc.measure(0,0)
qc.barrier()

for i in range(4):
    x = i%2
    qc.h(0).c_if(0, x)
    qc.measure(0, 0)

qc.draw('mpl')
```

```python
plot(qc)
```
