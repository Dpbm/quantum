```python
%matplotlib inline
from qiskit import QuantumCircuit, Aer, transpile
from qiskit.visualization import plot_histogram
```

```python
oracle = QuantumCircuit(6, name="oracle")
oracle.cx(0, 3)
oracle.cx(1, 4)
oracle.cx(2, 4)
oracle.cx(1, 5)
oracle.cx(2, 5)
oracle.draw('mpl')
```

```python
qc = QuantumCircuit(6,3)
qc.h(range(3))
qc.append(oracle.to_gate(), range(6))
qc.h(range(3))
qc.barrier()
qc.measure(range(3), range(3))
qc.draw('mpl')
```

```python
sim = Aer.get_backend('aer_simulator')
results = sim.run(transpile(qc, sim), shots=1000).result().get_counts()
plot_histogram(results)
```

$(b0 * 1) + (b1 * 1) + (b2 * 0) = 0$\

$(b0 * 1) + (b1 * 1) + (b2 * 1) = 0$\
$(b0 * 1) + (b1 * 1) + (0 * 1) = 0$\
