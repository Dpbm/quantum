```python
from qiskit import QuantumCircuit, Aer
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
```

```python
sim = Aer.get_backend('aer_simulator')
```

```python
qc = QuantumCircuit(1, 1)

qc.x(0)

qc.barrier()

qc.h(0)
qc.z(0)
qc.h(0)
qc.draw("mpl")
```

```python
state = Statevector(qc)
state.draw("latex")
```

```python
plot_bloch_multivector(state)
```

```python
z_measurement = QuantumCircuit(1, 1)
z_measurement.measure([0], [0])
z_measurement.draw("mpl")
```

```python
qc = qc.compose(z_measurement)
```

```python
qc.draw("mpl")
```

```python
results = sim.run(qc).result().get_counts()
results
```

```python
qc = QuantumCircuit(3, 3)

qc.x(0)

qc.barrier()


qc.h([0, 1])

qc.z([0, 1])

qc.h([0, 1])

qc.ccx(0, 1, 2)


qc.h([0, 1])

qc.z([0, 1])

qc.h([0, 1])

qc.barrier()

qc.draw("mpl")
```

```python
state = Statevector(qc)
state.draw("latex")
```

```python
plot_bloch_multivector(state)
```
